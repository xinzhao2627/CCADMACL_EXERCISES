```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x))
```

```python
df = list(client_basket.apply(lambda x:x.split(",")))
df
```

# One hot encoding


```python
from mlxtend.preprocessing import TransactionEncoder
from sklearn.preprocessing import OneHotEncoder


one_hot_transformer = TransactionEncoder()
df_transform = one_hot_transformer.fit_transform(df)
df_transform
```

```python
df = pd.DataFrame(df_transform,columns=one_hot_transformer.columns_)
df
```

# 2. Use the Apriori Algorithm to find the most frequent sets

```python
df_res =  apriori(df, min_support = 0.002, use_colnames = True)

```

```python
df_res
```

```python
df_res.sort_values(['support'],ascending=False, inplace = True)
```

```python
df_ar = association_rules(df_res, metric="lift", min_threshold=1)
```
